# Introduction to quantum computing and semiconductor qubits

## Why quantum computing? From classical to quantum bits

Modern classical computers are powered by billions of <abbr title="A transistor can be seen as an electronic switch that controls the flow of electricity. It is made by semiconductors and can turn circuits on and off and amplify weak signals.">transistors</abbr> that we can control to switch between clear-cut binary states – that is, either 0 (off) or 1 (on).
These *two-level systems* made by semiconductors are controllable via the application of a voltage and are the fundamental building blocks of modern computers and electronics.
We encode information as 0s and 1s implemented by transistors.

For decades, improvements in computing performance were mainly driven by shrinking the size of transistors ([Moore’s Law 🔗](https://en.wikipedia.org/wiki/Moore%27s_law)).
Today, their size reaches scales of only a [few nanometres 🔗](https://www.intel.com/content/dam/www/public/us/en/documents/corporate-information/history-moores-law-fun-facts-factsheet.pdf), where quantum effects such as electron tunnelling, energy quantization and others start to become relevant.
These effects create fundamental limits on further ‘classical scaling’.

Also, despite the improvements due to miniaturization, with denser central processing units (CPUs) operating at faster processing speeds (frequencies), classical computing has some important limitations.
There are certain classes of problems that shall remain out of the reach of classical computers if we want to solve them within reasonable time scales.
Although *classical algorithms* have benefited from decades of improvements and optimizations, they remain fundamentally limited for these types of problems, even if we use the fastest computers available, [some computations using them could take several years 🔗](https://comptes-rendus.academie-sciences.fr/physique/articles/10.5802/crphys.229/).

### Algorithm complexity: the quantum advantage

We can, in fact, define a metric to compare how fast different algorithms perform on a same computer.
In this case, the efficiency of an algorithm can be measured by how the number of steps needed to execute it (its *run time*) grows with the size of the task.
In simple terms, the size of the task can be measured by the number $ n $ of elements on which we need to execute our algorithm.
That is, $n$ is the size of the input.

We can then classify algorithms by how their run times scales.
For example, polynomial-like run times grow like $ n^x $ (where $ x > 0 $) whilst exponential run-times scale like $ e^n $, vide the figure below:

<div><img src="figures/fig_algo_time.py.png" width="1000px"/></div>

It is therefore clear that finding and using algorithms whose run times are polynomial-like with increasingly smaller exponents or logarithm-like is desirable. Unsurprisingly, intense research has been done into optimizing algorithms to reduce their [time complexity 🔗](https://en.wikipedia.org/wiki/Time_complexity).
Relevant for us is the existence of algorithms based on <abbr title="In quantum mechanics, we must move beyond the deterministic framework of classical physics and adopt a fundamentally probabilistic description of nature. While classical physics allows precise predictions of individual events given exact initial conditions, quantum mechanics instead provides statistical predictions for the outcomes of many repeated experiments. Each individual measurement (a single experiment) will yield a random result, but the probabilities governing such outcomes are precisely determined by the theory.">quantum mechanics</abbr> that offer speed-ups over the best known classical algorithms for certain problems.
In some cases, quantum algorithms reduce the degree of the polynomial runtime, while in others the improvement is more dramatic, transforming exponential-time scaling into polynomial-time scaling. 

Because these algorithms rely on the principles of quantum mechanics, they require **qubits**, the fundamental unit of quantum information, and ways to manipulate them, which we call *gates*.

### Qubit

Unlike a classical bit, which can be in a state of either 0 or 1, a qubit can be in a superposition of both states simultaneously.

Additionally, qubits can exhibit *entanglement*, a strong correlation between several of them that enables parallel computation on many states at once.
This property, combined with *superposition* and *interference*, allows quantum algorithms to achieve computational advantages unattainable by classical means alone.


Mathematically, if we represent the state of a single qubit as $ |\psi\rangle $, we can represent a general pure state and possibly superposed single-qubit state as

$\displaystyle
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$

where $\alpha$ and $\beta$ are complex probability amplitudes such that $|\alpha|^2 + |\beta|^2 = 1$.

These coefficients tell us how likely it is for the qubit, when measured, to be in the state $|0\rangle$ or the state $|1\rangle$.
Note that, indeed, when *measured*, a qubit will collapse to either $|0\rangle$ or $|1\rangle$ with a probability given by $|\alpha|^2$ and $|\beta|^2$, respectively.

It is useful to represent a single qubit geometrically using the Bloch sphere:

<div><img src="figures/fig_bloch_sphere.png" width="1000px"/></div>

It is a useful visual model where any qubit state corresponds to a point on the surface of a sphere and single-qubit gates correspond to rotations on this sphere.

In a Bloch sphere, the north pole is the $|0\rangle$ state, the south pole is the $|1\rangle$ state and all other points are superpositions. Also,
- $\theta$ is the polar angle, which can be associated with the degree of superposition, and
- $\varphi$ is the azimuthal angle, which can be related to the quantum phase.

Mathematically, a general single-qubit state can then be written in terms of these variables as

$\displaystyle
|\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + \mathrm{e}^{\mathrm{i}\varphi}\sin\left(\frac{\theta}{2}\right) |1\rangle.
$

### Gates

Classical gates, which are simply binary operations, are the fundamental building block of digital circuits and are used to perform basic logical operations on binary inputs.
They process signals representing 0s and 1s to produce a single binary output based on Boolean logic and an algorithm can be constructed by combining several gates.

Some examples are the NOT, AND and OR gates. Their symbols and truth tables are

<div><img src="figures/fig_gates.png" width="500px"/></div>

A gate in quantum computing is a fundamental operation that manipulates the state of a collection of qubits, analogous to logic gates in classical computing. Unlike classical gates, however, the quantum ones may operate on superpositions, enabling transformations like rotation, entanglement and others, in addition to being reversible (mathematically, they are unitary transformations).

Common examples of single-qubit quantum gates include:
- X gate: Analogous to a classical NOT gate. That is, it turns $|0\rangle$ into $|1\rangle$ and vice versa.
- Hadamard gate: Creates superposition from a basis state. For instance by applying this gate on the state $|0\rangle$ the result will be $|0\rangle + |1\rangle)/\sqrt{2}$.

Their effect can also be represented in the Bloch sphere. If we consider a starting state at $|0\rangle$:

The two examples described above are known as single-qubit gates, since they act on a single qubit.
Naturally, there are other quantum gates can change the state of a two-qubit system.

#### CNOT Gate Example

One example of a two-qubit gate is the **CNOT gate**.
In a pure two-qubit system, a general state can be written as $\ket{x}\ket{y} = \ket{x,y}$, where $x$ and $y$ are 0 or 1.
The CNOT gate acts as  

$\text{CNOT} \ket{x,y} = \ket{x, x \oplus y}$,

where $\oplus$ denotes addition modulo 2. In words, the CNOT flips the target qubit if the control qubit is $\ket{1}$ and does nothing if the control qubit is $\ket{0}$.

The truth tables for the XOR and CNOT gates are

It is clear that the CNOT gate is analogous to a classical XOR operation, but there are important differences: 

- The CNOT gate is reversible: the control qubit is preserved and, as such, the operation can be undone.  
- In contrast, a classical XOR gate does not always retain enough information to be reversible.  
- The CNOT can generate [quantum entanglement 🔗](https://en.wikipedia.org/wiki/Quantum_entanglement), a purely quantum correlation effect with no classical analogue.  

Naturally, as in the classical case, by combining several quantum gates, one can implement quantum algorithms.

## Platforms for qubits: electron spins in silicon quantum dots

The most common classical two-level system that give us the bit is the transistor.
For for a qubit there are several possible ‘hardware’ implementations capable of realizing them.
In all of them, we should be able to isolate two energy levels to use them as the basis (our 0s and 1s) for our qubits and we should be able to manipulate these states as to perform quantum gates.

Among others, some of these qubit platforms include:

* Superconducting circuits: Small circuits made with special materials in which electrons flow without resistance when cooled to near absolute zero and maintain quantum coherence over macroscopic scales.
  The dimensions of a superconducting qubit are around 0.01 cm ($10^{-4}$ m).
* Trapped ions: Individual ions are confined in vacuum by electromagnetic fields and manipulated using laser pulses.
  Each qubit is encoded in the internal electronic state of a single ion, giving it an effective size comparable to that of an atom (on the order of $10^{-10}$ m).
  The ions themselves are held at fixed positions in the trap, typically separated by distances on the order of $10^{-6}$ m.

and also the one which shall be the focus of this course:

* Semiconductor (spin) qubits: Tiny electronic devices, similar to transistors, made from materials like [silicon 🔗](https://spj.science.org/doi/10.34133/icomputing.0115) or [germanium 🔗](https://qutech.nl/2025/07/09/quantum-engineering-of-germanium-for-scaling-quieter-qubits/), which can store and process quantum information using the properties of electron spins.
  The dimensions of a spin qubit are around 0.00001 cm ($10^{-7}$ m).
  Its fundamental building block is the electron spin, with the most relevant system for quantum computing being electron spins confined in silicon quantum dots.

Each of these and other qubit platforms involves a complex surrounding engineering system – which may include cryogenics, ultra-high vacuum, precision lasers and others – to create and control the different qubits.
Moreover, one of the biggest issues to overcome to improve quantum computers is related to filtering unwanted disturbances from the environment (which we call *noise*) that may cause qubits to lose the quantum information registered on them, making calculations unreliable.
This requires improvements on the control system surrounding the qubits.

These limitations can be mitigated by encoding information across multiple physical qubits to form a single logical qubit.
Whilst individual physical qubits are inherently imperfect and susceptible to noise, decoherence and operational errors, collective encoding enables error correction and results in logical qubits with improved stability and reliability. 

Performing useful quantum computations therefore requires not only many physical qubits per logical qubit, but also the ability to realize large numbers of logical qubits that are protected by error correction.
This places strong demands on hardware scalability and control, turning quantum computing into a major engineering challenge.
As a result, multiple physical platforms are currently being explored.
And it remains an open question which technology will ultimately provide the best combination of qubit count, controllability and operational fidelity, much as silicon-based transistors emerged as the dominant platform in classical computing.

## Overview of concepts

### Electric field and potential

An **electric field** can be understood as a field defined at every point in space that determines the force experienced by charged particles (electrons, protons and so on) or objects.

The field will exert a force on other charges within that region, either attractive (if the charges have opposite signs) or repulsive (if the charges have the same sign).
This is similar to a magnet’s influence, but for electric charges. The strength and direction of this force at each point in space define the field.

The **electric force** $\vec{F}$ and field $\vec{E}$ are related via the **charge** $q$ of a given particle as follows:

$\displaystyle
\vec{F} = q\vec{E} \implies \vec{E} = \frac{\vec{F}}{q}
$

Another key quantity is the **electric potential** $\varphi$.
It measures how much electric potential energy a unit charge would have at a specific point in an electric field.
It plays a role similar to ‘height’ in gravitational systems: just as an object at a higher elevation point has more gravitational potential energy, a charge at a point of higher electric potential has more electric potential energy.

The electric potential $\varphi(\vec{r})$, which generally varies in function of the position $\vec{r}$, is related to the electronic charge density $\rho(\vec{r})$ by [Poisson’s equation 🔗](https://en.wikipedia.org/wiki/Poisson%27s_equation),

$\displaystyle
\nabla^2 \varphi(\vec{r}) = -\frac{\rho(\vec{r})}{\epsilon},
$

where $\rho$ is charge density and $\epsilon$ is the [permittivity of the material 🔗](https://en.wikipedia.org/wiki/Permittivity). Also, $\nabla^2 \varphi(\vec{r})$ is the rate of change of the electric potential over the spatial coordinates $x$, $y$ and $z$.

Lastly, **voltage** is simply the difference in electric potential between two points in an electric field. If electric potential tells you how much potential energy a charge has at a single point, voltage tells you how much energy changes between two points.

In this course, the devices studied will be designed and tuned to find the best values for $\varphi$ as a function of position, $\vec{r}$.

### Wavefunctions and quantum states

When the size of physical systems becomes sufficiently small, we must describe their behaviour and that of their constituents using **quantum mechanics**, which provides the foundational rules governing microscopic systems.
This framework introduces fundamental principles such as **superposition**, **entanglement** and the **probabilistic nature** of measurements, all of those properties being fundamental to quantum computing.
Consequently, probability becomes an inherent and essential aspect of modelling phenomena at the quantum scale.

Mathematically, in quantum mechanics, each object is fully described by a function of position $\vec{r}$ and time $t$ called the **wave function** and denoted $\Psi(\vec{r}, t)$ (which is associated to the vector state $|\Psi\rangle$).
The wave function in general is a complex quantity and the product of its complex conjugate with itself, $\Psi^*(\vec{r},t)\Psi(\vec{r},t)$ gives the probability density of finding the particle at that position and time.

To find the wave function $\Psi$, we need to solve [Schrödinger’s equation 🔗](https://en.wikipedia.org/wiki/Schr%C3%B6dinger_equation).
It governs the time evolution of quantum systems and determines the allowable energy states in confined systems.
Some examples of confined systems are the electron bound to a proton, forming a hydrogen atom, and electrons in the valence band of semiconductors, which can only have specific energies.

Relevant for us, in the context of electrostatic simulations for a particle of mass $m$, Schrödinger’s equation can be written as

$\displaystyle
\mathrm{i}\hbar \frac{\partial}{\partial t} \Psi(\vec{r}, t) =
-\frac{\hbar^2}{2m}\nabla^2 \Psi(\vec{r}, t) + V(\vec{r})\Psi(\vec{r}, t),
$

where $\mathrm{i}=\sqrt{-1}$, $\hbar$ is the reduced Planck’s constant, $(\partial/\partial t)\Psi(\vec{r}, t)$ is the time derivative of the wave function (the rate of its change over time) and $\nabla^2 \Psi(\vec{r}, t)$ is the rate of the rate of change of the wave function over the spatial coordinates ($\vec{r} = x \hat{x} + y \hat{y} + z\hat{z}$).
Furthermore, $V(\vec{r})$ is the total confinement potential experienced by the particle.
For electrons, it is related to the electric potential $V(\vec{r}) = e\varphi(\vec{r})$, where $e$ is the charge of the electron.

In many cases, including the electrostatic simulations we perform in QTCAD®, the temporal and spatial parts of the wave function can be considered separable:

$\displaystyle
\Psi(\vec{r}, t) = \Psi_r(\vec{r}) \Psi_t(t)
$

Under this assumption, Schrödinger’s equation can be reduced to its time-independent form:

$\displaystyle
-\frac{\hbar^2}{2m}\nabla^2 \Psi_r(\vec{r}) + \varphi(\vec{r})\Psi_r(\vec{r}) = E \Psi_r(\vec{r}),
$

where $E$ is the energy eigenvalue associated with the stationary state $\Psi_r(\vec{r})$. This form is particularly convenient for solving problems involving electrons or holes confined in quantum dots, where the potential $\varphi(\vec{r})$ is determined by the device geometry and the applied gate voltages.

Modifying the electric potential $\varphi(\vec{r})$ will significantly influence the system’s quantum dynamics and the resulting solutions for $\Psi_r(\vec{r})$.
In fact, in this course, we will solve Schrödinger’s equation and study the effects of having different potentials $\varphi(\vec{r})$.

### Energy levels and bands

As mentioned, in quantum mechanics, electrons in a material (which can be the electron in a hydrogen atom, for instance) cannot have arbitrary energies.
Instead, they are restricted to specific discrete **energy levels** (or states) which are unevenly separated from each other.

Moreover, when electrons are added to the system (the material), they start by filling (that is, occupying) the states with the lowest possible energy.
Importantly, if we ignore spin, only a single electron can occupy a given discrete energy state.
Once we consider its spin, if the spin of the electrons are different, up to two electrons can have the same energy, that is, occupy the same discrete energy level.
Such a discretization of the energy levels (compared to an electron not bound to any atom or, for all purposes, a classical system) is a direct consequence of the quantum nature of electrons and the constraints imposed by their environment.

The second column shows the levels when we include relativistic corrections and the spin of the electron; this is the fine structure.
The third column shows how the levels change when we include effects from the nucleus; this is the hyperfine structure.

**Spin**

The **spin** is an intrinsic form of [angular momentum 🔗](https://en.wikipedia.org/wiki/Angular_momentum) carried by elementary particles such as electrons.
Unlike orbital angular momentum, which arises from the rotational motion of particles in space, spin is a fundamental quantum property with no classical analogue.
It has a fixed magnitude and can exist in **two discrete states**.
We call those ‘spin up’ ($|\uparrow\rangle$) and ‘spin down’ ($|\downarrow\rangle$) and the mathematics that describe such a two-level system can be used to explain other similar systems.

Furthermore, the spin of a particle interacts with magnetic fields and leads to energy splitting.
In the absence of a magnetic field $\vec{B}$, the energies $E$ of bound electrons with in the spin up and down states are the same, $E_{|\uparrow\rangle} = E_{|\downarrow\rangle}$.
This is not the case when there is a magnetic field.
If it is finite, the field will split the spin states by the Zeeman energy such that $E_{|\uparrow\rangle} - E_{|\downarrow\rangle} = g\mu_\mathrm{B} |\vec{B}|$, where $g$ is the [electron gyromagnetic factor 🔗](https://en.wikipedia.org/wiki/G-factor_(physics)#Electron_g-factors) and $\mu_\mathrm{B}$ is the [Bohr magneton constant 🔗](https://en.wikipedia.org/wiki/Bohr_magneton).

As a matter of fact, energy levels split due to the electron spin are optimal systems for quantum computing; taken in isolation, they work as an *artificial* atom with two levels: a qubit.

</div>

In an isolated atom, as in the example above, electrons occupy sharply defined <abbr title="The atomic orbitals are common labelled as 1s, 2p et cætera.">orbitals each with a precise energy</abbr> and mostly distributed in a small region of space.
However, when atoms are brought together to form a solid, their orbitals overlap and interact.
The discrete atomic levels then become broad, forming **energy bands**.
These bands are continuous ranges of energies allowed for the electrons, which are separated by band gaps.
These band gaps are energies for which no energy states are available.
Some finite energy must be given to the electrons for them to become excited and jump between the different bands, leaving behind positively charged quasi-particles called ‘holes’.

The concept of energy bands is fundamental to describe semiconductors, which we will overview next.

### Semiconductors

Semiconductors are materials with electrical conductivity between that of conductors and insulators.
Their unique property is that conductivity (the mobility of electrons that can particiate in the transport of electric current) can be precisely controlled by factors like temperature, impurities (doping) and, more importantly for our case, electric fields.

This tunability arises from their electronic band structure, where a **small energy gap** – around $0.1$–$2$ electronvolts (eV) – separates the valence band (filled with electrons) from the conduction band (empty at low temperatures and responsible for conduction of electric current when it is populated).
Adding donor or acceptor atoms via doping introduces extra electrons or holes, enabling the fine control of electric current, which is essential for modern electronics.

<div><img src="figures/fig_semiconductor_comparison.png" width="1000px"/></div>

At finite temperatures or under external influence (electric field, for instance), electrons can jump from the valence band to the conduction band.
Since we can precisely tune these external parameters, semiconductors allow for controllable electrical conduction.

In fact, [transistors 🔗](https://en.wikipedia.org/wiki/Transistor) leverage exactly this semiconductor behaviour.
Typically made from silicon (a semiconductor), transistors act as switches or amplifiers by regulating current between two terminals via a third control terminal.
By applying an electric potential difference (a voltage) to the control gate, the conductivity of the semiconductor can be changed, allowing binary off or on states (0 or 1, respectively) to represent and process information.

This precise control over charge in semiconductors via electric fields not only powers classical devices but also provides the basis for quantum bits (qubits).
Relevant to us is the case of spin qubits in quantum dots, where the quantum states of electrons (or holes) in semiconductors encode information.

### Quantum dots

A quantum dot is a nanoscale semiconductor device that confines electrons in all three spatial dimensions.
This confinement reintroduces discrete energy levels within the material’s continuous bands, leading to an artificial atom.
Electrons may occupy these discrete levels and the transitions between these can be precisely controlled with electric or magnetic fields, which makes quantum dots suitable platforms for qubits.

To create a prototypical quantum dot, we need a layered structure like the following:
<div><img src="figures/fig_quantum_dot_simple.png" width="400px"/></div>

Here, the semiconductor layer will ‘host’ the quantum dot.
One also places an insulating (oxide) layer to isolate the semiconductor layer from the metallic gates placed on top of the oxide layer.
These gates, together with the bottom one, form a control system.

By tuning the voltages of these gates, we can create an electric potential $\varphi(\vec{r})$ with a specific profile (depth and shape) around a small region of the semiconductor layer that will trap a single electron in three dimensions.
This is called the **confinement potential**.
<div><img src="figures/fig_quantum_dot.png" width="1000px"/></div>

The possible energy levels inside the quantum dot region are obtained by solving the Schrödinger equation with the electric potential $\varphi(\vec{r})$ generated by the gate.
Also, to ensure the electron remains confined within the dot region, we need to impose appropriate boundary conditions on the behaviour of the wave function.
We will then find energy levels and their values, analogous to atomic orbitals (like in the hydrogen case discussed above), but tunable via **gate voltages**.
That is, we can tune the energy levels by changing the voltages applied on the gates of the device.

Furthermore, to use a quantum dot as a qubit, we must also add additional gates to the semiconductor device to provide a ‘source’ of electrons to enter the quantum dot region and a ‘drain’ for them (electrons exiting the quantum dot region).
By adjusting the gate voltage, we can then shift the energy levels of the quantum dot to allow or block the passage of electrons between the source and the drain.
